```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

# T5
- Encoder-decoder model
- Encoder + causal decoder attention
- Usage: Text-to-text mapping
- Mask:
+ T5 encoder: padding mask -> source token attend bidirectionally
+ T5 decoder: causal mask + cross attention mask -> target attends left to right and read source.
## Core idea
- Topo:
+ X_src -> Encoder -> H_enc
+ Y<t, H_enc -> Decoder -> p(y_t)
- Unified text-to-text view: Every task is formulated as: text input -> text output
- T5 is natural for tasks with explicit source-to-target transformation
## Attention
T5 decoder both remembers generated prefix and read encoded source
### Encoder self-attention
Bidirectional attention over source tokens
### Decoder self-attention
Causal attention over generated target tokens
### Cross attention
Decoder queries attend to encoder keys/values: Q_dec K_enc ^T
## Training objective
### Seq2seq
Given input x, predict output text y_1, ..., y_m. with p(y_t | y<t, x)

```python
ckpt = 't5-small'
tok = AutoTokenizer.from_pretrained(ckpt)
model = AutoModelForSeq2SeqLM.from_pretrained(ckpt)

text = "translate English to German: The house is wonderful."
inputs = tok(text, return_tensors='pt')

outputs = model.generate(**inputs, max_new_tokens=40)
print(tok.decode(outputs[0], skip_special_tokens=True))
```

## Usage
- Machine translation
- Summarization
- Structured generation from input text
- Text normalization and rewriting
- Tasks where output length differs greatly from input length
