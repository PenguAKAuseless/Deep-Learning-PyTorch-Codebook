```python
import torch
import torch.nn as nn
```

# Pooling
- Reduce spatial resolution, summarize local regions
- Position: Conv -> Norm -> Activation -> **Pool** -> Dropout
- Do not preserve the shape
- Do not learn (nonlinear transformation)
- y_b,c,i,j = pool(x_b,c,local-window)
- Shape: Out = (In + 2 * Pad - Kernel) / Stride + 1 (Convolution formula)

```python
# Keep strongest local response
# Commonly used after ReLU
# Feature presence > location
class MaxPooling2D(nn.Module):
    def __init__(self):
        super().__init__()
        self.max_pooling = nn.MaxPool2d(kernel_size=2, stride=2)

    def forward(self, x):
        return self.max_pooling(x)
```

```python
# Smooth local response
# Commonly used at the end of the network as global average pooling
# Average presence > peak evidence
class AveragePooling2D(nn.Module):
    def __init__(self):
        super().__init__()
        self.avg_pooling = nn.AvgPool2d(kernel_size=2, stride=2)

    def forward(self, x):
        return self.avg_pooling(x)
```
