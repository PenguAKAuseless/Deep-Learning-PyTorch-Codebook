```python
import torch
import torch.nn as nn
```

# Linear
B x D in
B x K out
Parameter count: D x K + K (if bias)

# Activation
Binary: Sigmoid
Single label: Softmax
Multi label: Sigmoid over class
No parameter required

```python
# Linear Regressor example
class LinearRegressor(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()
        self.linear = nn.Linear(in_features, out_features)
    
    def forward(self, x):
        return self.linear(x)
    
model = LinearRegressor(10, 1)
criterion = nn.MSELoss()
```

```python
# Linear Classifier example
class LinearClassifier(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()
        self.linear = nn.Linear(in_features, out_features)
    
    def forward(self, x):
        return self.linear(x)

model = LinearClassifier(in_features=10, out_features=3)
criterion = nn.CrossEntropyLoss() # BCEWithLogitsLoss() for binary classification and multi-label classification
```
