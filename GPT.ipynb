```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

# GPT
- Decoder-only model
- Causal attention
- Generation, instruction following
- Mask: causal + padding mask -> token can't attend to future tokens
## Core idea
- Topo: X<=t -> causal decoder stack -> p(x_t+1 | x<=t)
- Attention: causal self-attention (unidirectional)
- Designed for autoregressive generation
## Traing objective
### Causal Language Modeling
- Given: x_1, x_2, ..., x_n
- Predict: p(x_t | x<t)
### Loss
L = - sum(log p(x_t | x<t))
## Inference - Autoregressive decoding
Output quality depends on both model and decoding strategy
### Generation loop
x_1, x_2, ..., x_t -> p(x_t+1) -> x_t+1 -> p(x_t+2) -> ...
### Decoding choices
- Greedy decoding
- Beam search
- Top-k sampling
- Nucleus/top-p sampling
- Temperature scaling
## GPT and KV cache
- Solve issue recomputes attention over all previous tokens at every step
- Store keys and values from previous tokens: K<=t, V<=t
- Compute only new Q K V for new token
- KV cache is essential for efficient GPT inference

```python
ckpt = 'gpt2'
tok = AutoTokenizer.from_pretrained(ckpt)
model = AutoModelForCausalLM.from_pretrained(ckpt)
inputs = tok('Hello, how are you?', return_tensors='pt')
outputs = model.generate(
    **inputs, 
    max_new_tokens=30,
    do_sample=True,
    top_p=0.95,
    temperature=0.8,
)
```

## Usage
- Open-ended text generation
- Chatbots and instruction following
- Code generation
- Summarization and QA via prompting
- In-context learning
- IMPORTANT: GPT-style models can generate fluent but incorrect text (hallucinate) -> need additional system design for factuality and grounding
