# Vision Transformer (ViT-Base/16)
Patch Embedding -> Transformer Encoder (12 Blocks) -> Classification Head
- Kích thước ảnh chuẩn: 224x224.
- Kích thước Patch (Patch Size): 16x16.
- Embedding Dimension (D): 768.
- Input: 224 x 224 -> Patch: 16 x 16

## 1. Input
`B x C x H x W = B x 3 x 224 x 224`

## 2. Patch Embedding & Tokens
Conv2D(768, 3, 16, 16, 0) -> Flatten(2) -> Transpose (1, 2)
- Patch count = 224 / 16 * 224 / 16 = 196
- Output `B x 768 x 14 x 14`.
- Flatten dimension 2: `B x 768 x 196`. Each patch now is a 768 dimension vector (-> Transpose(1, 2)) 
- **Class Token (CLS):** Added to start of sequence to learn representation `B x 197 x 768`.
- **Positional Embedding:** Added learnable Positional Encoding (`1 x 197 x 768`) into tensor
