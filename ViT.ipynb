# ViT
- Encoder-only model
- Bidirectional attention over patches
- Usage: Image classification, vision backbone
- No mask -> image patches attend globally
## Core idea
- Tokenization: split image into patches
+ X (C, H, W) -> N patches
+ N = HW/P^2
- Patch embedding: Each patch projected to d_model
- Convert an image into a token sequence
- Patch tokens: X_patch (B, N, d)
- CLS token: X_vit (B, N + 1, d)
- Position embedding: X_0 = X_vit + E_pos
## Compare with CNN
- ViT trade inductive bias for flexible global attention
- Token unit: image patches (CNN use pixels/local feature maps)
- Benefits from large-scale pretraining
