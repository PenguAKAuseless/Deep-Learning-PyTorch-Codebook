# Dropout
- Regularization technique, do not change shape
- Randomly set activations to zero
- p = probability dropping activation
- Usually Y = M dot X (element wise multiplication)
- Modern framework: Y = (M dot X) / (1 - p) (preserve activation magnitude, no extra scaling at eval)
- Train: Random mask is sampled, set some activations to 0, scale other activations
- Eval: disabled, become y = x identity layer.
- Do not learn (Parameters = 0)
- Position: Conv -> Norm -> Activation -> Pool -> Dropout
- Put after FC, Conv; before head (avoid too heavy dropout, may hurt training)

```python
import torch
import torch.nn as nn

# Standard Dropout: Randomly zeroes some of the elements of the input tensor with probability p using samples from a Bernoulli distribution.# @# ###
class Dropout(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.dropout = nn.Dropout(p)

    def forward(self, x):
        return self.dropout(x)
```

```python
import torch
import torch.nn as nn

# Spatial dropout: randomly zeroes out entire channels in the input feature map.
class Dropout2D(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.dropout2d = nn.Dropout2d(p)

    def forward(self, x):
        return self.dropout2d(x)
```
