# Dense Prediction
- Idea: Encoder (Conv2D stack) -> Get semantic features. Decode: full resolution, assign a label to each pixel -> Upsampling
- Combine with skip connection
- In: B x C_in x H_in x W_in
- Out: B x C_out x H_out x W_out
- H_out >= H_in when s > 1 (padded)
- Out = (In - 1) * Stride + Kernel - 2 * Pad + OutPad
- Parameters: C_out * C_in * H_K * W_K + C_out

```python
import torch
import torch.nn as nn

class ConvTranspose2D(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv_transpose = nn.ConvTranspose2d(in_channels=16, out_channels=3, kernel_size=3, stride=1, padding=1, bias=True)

    def forward(self, x):
        return self.conv_transpose(x)
```
